```python
# line chart showing attributed patients over time

# Convert 'attributed_start_date' column to datetime format
patient_df['attributed_start_date'] = pd.to_datetime(patient_df['attributed_start_date'])

# Specify the desired start date
start_date = pd.to_datetime('2023-01-01')

# Filter the DataFrame to include only data starting from the desired start date
patient_df_filtered = patient_df[patient_df['attributed_start_date'] >= start_date]

# Create a new column with month and year
patient_df_filtered['month_year'] = patient_df_filtered['attributed_start_date'].dt.strftime('%Y-%m')

# Group by month and year and count the number of patients for each month
patients_per_date = patient_df_filtered.groupby('month_year').size().reset_index(name='patient_count')

# Calculate cumulative sum of patients over time
patients_per_date['cumulative_patients'] = patients_per_date['patient_count'].cumsum()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(patients_per_date["month_year"], patients_per_date['cumulative_patients'], marker='o', linestyle='-')
plt.title('Total number of patients in the program')
plt.xlabel('Year & month')
plt.ylabel('Total number of patients')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
age_groups = {
    '0-4': ['0', 'Less than 1', '1 to 4'],
    '5-19': ['5 to 9', '10 to 14', '15 to 19'],
    '20-39': ['20 to 24', '25 to 29', '30 to 34', '35 to 39'],
    '40-59': ['40 to 44', '45 to 49', '50 to 54', '55 to 59'],
    '60+': ['60 to 64', '65 to 69', '70 to 74', '75 to 79', '80+']
}

age_mapping = {
    '0': '0-4', 'Less than 1': '0-4', '1 to 4': '0-4',
    '5 to 9': '5-19', '10 to 14': '5-19', '15 to 19': '5-19',
    '20 to 24': '20-39', '25 to 29': '20-39', '30 to 34': '20-39', '35 to 39': '20-39',
    '40 to 44': '40-59', '45 to 49': '40-59', '50 to 54': '40-59', '55 to 59': '40-59',
    '60 to 64': '60+', '65 to 69': '60+', '70 to 74': '60+', '75 to 79': '60+', '80+': '60+'
}

# Map age bands to age groups and add as a new column to the original DataFrame
patient_df['age_group'] = patient_df['age_band'].map(age_mapping)
```

```python
category_order = ['0-4', '5-19', '20-39', '40-59', '60+']

sns.countplot(x='age_group', data=patient_df, order=category_order, palette=qualitative_colors)

plt.title('Number of patients by age group')
plt.xlabel('Age group')
plt.ylabel('Total number of patients')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
#plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
filtered_risk_df = patient_df[~patient_df['major_segment'].isin(["not known", "Not active on scheme"])]
segment_age_gender_counts = filtered_risk_df.groupby(['major_segment', 'age_group']).size().reset_index(name='count')
display(segment_age_gender_counts)
```

```python
desired_order_rows = ['Healthy', 'At Risk', 'Maternity', 'Sig Condition', 'High risk']
desired_order_columns = ['0-4', '5-19', '20-39', '40-59', '60+']

pivot_df = filtered_risk_df.pivot_table(index='major_segment', columns='age_group', values='gender', aggfunc='count')

# Reorder rows and columns
pivot_df = pivot_df.reindex(index=desired_order_rows, columns=desired_order_columns)
pivot_df.fillna(0, inplace=True)
pivot_df = pivot_df.round(0)

# Display the pivot table
display(pivot_df)
```

```python
desired_order_rows = ['Healthy', 'At Risk', 'Maternity', 'Sig Condition', 'High risk']
desired_order_columns = ['0-4', '5-19', '20-39', '40-59', '60+']

# Create the pivot table with counts
pivot_df = filtered_risk_df.pivot_table(index='major_segment', columns='age_group', values='gender', aggfunc='count')

# Reorder rows and columns
pivot_df = pivot_df.reindex(index=desired_order_rows, columns=desired_order_columns)

# Convert counts to percentages
pivot_df_percentages = round(pivot_df.div(pivot_df.sum(axis=1), axis=0) * 100,0)  # Divide by row sum and multiply by 100

# Create a formatting function
def format_percentage(value):
    return '{:.0f}%'.format(value)

# Apply formatting function to each cell
pivot_df_percentages_formatted = pivot_df_percentages.applymap(format_percentage)
pivot_df_percentages_formatted.replace('nan%', '0%', inplace=True)

#pivot_df_percentages_formatted['Total %'] = pivot_df_percentages.sum(axis=1)

# Display the formatted pivot table with the total percentage column
display(pivot_df_percentages_formatted)
```

```python
patient_df=patient_df.sort_values(by="attributed_start_date")

#cumulative sum of patients by year&month

patient_df['cumulative_patients'] = patient_df.groupby('attributed_start_date').cumcount() + 1

filtered_patient_df = patient_df.dropna(subset=['cumulative_patients'])

display(filtered_patient_df['cumulative_patients'])
display(filtered_patient_df['cumulative_patients'].isnull().sum())
```

```python
temp_df = filtered_patient_df.sort_values(by='attributed_start_date').groupby('attributed_start_date')[['cumulative_patients']].count().reset_index()

plt.plot(temp_df["attributed_start_date"], temp_df['cumulative_patients'])

# Set the labels and title
plt.xlabel('Date')
plt.ylabel('Cumulative Number of Patients')
plt.title('Cumulative Number of Patients Over Time')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
temp_df = filtered_patient_df.sort_values(by='attributed_start_date').groupby('attributed_start_date')[['number_of_patients']].sum().reset_index()
temp_df = temp_df.rename(columns={"number_of_patients":"cumulative_patients"})
# df.to_csv(), df.to_excel()

#sns.lineplot(data=temp_df, x="attributed_start_date", y="number_of_patients");
# Plot the cumulative number of patients over time; "you're"
plt.plot(temp_df["attributed_start_date"], temp_df['cumulative_patients'])

# Set the labels and title
plt.xlabel('Date')
plt.ylabel('Cumulative Number of Patients')
plt.title('Cumulative Number of Patients Over Time')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()

#plt.savefig("../images/figure.png")
# del df
```

```python
filtered_patient_df = filtered_patient_df[filtered_patient_df["attributed_start_date"].notnull()].copy()

# Create a new column for cumulative patient count
filtered_patient_df['cumulative_patients'] = filtered_patient_df['number_of_patients'].cumsum()

# Plot the cumulative number of patients over time
plt.plot(filtered_patient_df["attributed_start_date"], filtered_patient_df['cumulative_patients'])

# Set the labels and title
plt.xlabel('Date')
plt.ylabel('Cumulative Number of Patients')
plt.title('Cumulative Number of Patients Over Time')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
age_groups = {
    '0-4': ['0', 'Less than 1', '1 to 4'],
    '5-19': ['5 to 9', '10 to 14', '15 to 19'],
    '20-39': ['20 to 24', '25 to 29', '30 to 34', '35 to 39'],
    '40-59': ['40 to 44', '45 to 49', '50 to 54', '55 to 59'],
    '60+': ['60 to 64', '65 to 69', '70 to 74', '75 to 79', '80+']
}

age_mapping = {
    '0': '0-4', 'Less than 1': '0-4', '1 to 4': '0-4',
    '5 to 9': '5-19', '10 to 14': '5-19', '15 to 19': '5-19',
    '20 to 24': '20-39', '25 to 29': '20-39', '30 to 34': '20-39', '35 to 39': '20-39',
    '40 to 44': '40-59', '45 to 49': '40-59', '50 to 54': '40-59', '55 to 59': '40-59',
    '60 to 64': '60+', '65 to 69': '60+', '70 to 74': '60+', '75 to 79': '60+', '80+': '60+'
}

# Map age bands to age groups and add as a new column to the original DataFrame
patient_df['age_group'] = patient_df['age_band'].map(age_mapping)
```

```python
category_order = ['0-4', '5-19', '20-39', '40-59', '60+']

sns.countplot(x='age_group', data=patient_df, order=category_order)
plt.xticks(rotation=90)
plt.show()
```
